```python
import pandas as pd
import numpy as np
```

Using the Path library to read the file

```python
from pathlib import Path

project_root = Path.cwd().parent.parent
data_path = project_root / "data" / "raw" / "emails.csv"

df_em = pd.read_csv(data_path, header=0)
df_em.head(10)
```

This dataset contains email interaction data with customers, tracking metrics such as accreditation completion, sentiment, contractor engagement, and various complaints or concerns discussed.
Like the renewal calls, these features will help predict whether customers are at risk of churning after their subscription due date.

Drop columns that are completely null or have unnamed index formats.

```python
# Drop if every value in the column is missing
df_em = df_em.dropna(axis=1, how='all')

# Remove Unnamed columns
df_em = df_em.loc[:, ~df_em.columns.str.contains('^Unnamed')]

print(f"Columns remaining: {df_em.columns.tolist()}")
```

Check for null values for every column

```python
print(df_em.isnull().sum())
```

Checking unique values for each column to understand how the data looks before cleaning

```python
for col in df_em.columns:
    if col not in ['Co_Ref']:
        print(f"\n--- {col} ---")
        print(df_em[col].value_counts(dropna=False).head(10))
```

Handling NULL values for categorical and flag columns

```python
# The data contains columns that have 'Yes', 'No', and 'Not Discussed' 
# We fill missing values with 'Not Discussed' where appropriate.
cat_cols = df_em.select_dtypes(include=['object']).columns

for col in cat_cols:
    if col not in ['Co_Ref', 'Time_to_Renewal']:
        df_em[col] = df_em[col].fillna('Not Discussed')
```

Cleaning flag columns (Yes/No type) - standardizing free-text comments, stray numbers, and inconsistent entries into proper Yes/No/Not Discussed flags.

```python
flag_cols = [
    'crm_accreditation_completed', 'crm_timely_completion', 'crm_progress_towards_accreditation',
    'crm_delays_in_accreditation', 'crm_contractor_suggested_leave', 'crm_contractor_engagement',
    'crm_dts_or_ssip_mentioned', 'crm_customer_payment_intention', 'crm_competitors_mentioned',
    'crm_platform_issues_raised', 'crm_agent_chased_contractor', 'crm_accreditation_issues',
    'crm_membership_overdue', 'crm_dissatisified_with_renewal_price', 'crm_customer_complained',
    'crm_refund_mentioned', 'crm_negative_customer_experience', 'crm_dissatisfaction_with_support',
    'crm_financial_hardship_mentioned'
]

valid_flags = ['Yes', 'No', 'Not Discussed']

for col in flag_cols:
    if col in df_em.columns:
        df_em[col] = df_em[col].astype(str).str.strip()
        # Map '0' to 'No'
        df_em[col] = df_em[col].replace({'0': 'No'})
        # Anything that is not a valid flag is a free-text comment, map to 'Yes'
        invalid_mask = ~df_em[col].isin(valid_flags)
        df_em.loc[invalid_mask, col] = 'Yes'
```

Verify flag columns now only contain Yes / No / Not Discussed

```python
for col in flag_cols:
    if col in df_em.columns:
        print(f"{col}: {df_em[col].unique()}")
```

Cleaning the crm_contractor_sentiment column - it should only have Neutral, Satisfied, Dissatisfied, or Not Discussed

```python
print("Before cleaning:")
print(df_em['crm_contractor_sentiment'].value_counts())
```

```python
# 'Yes' and 'No' do not belong in a sentiment column, map them to 'Not Discussed'
valid_sentiments = ['Neutral', 'Satisfied', 'Dissatisfied', 'Not Discussed']
df_em['crm_contractor_sentiment'] = df_em['crm_contractor_sentiment'].apply(
    lambda x: x if x in valid_sentiments else 'Not Discussed'
)

print("After cleaning:")
print(df_em['crm_contractor_sentiment'].value_counts())
```

Cleaning the crm_contractor_sentiment_score column - converting to numeric, removing text labels like 'Dissatisfied', 'Neutral', 'Yes', etc.

```python
print("Before cleaning:")
print(df_em['crm_contractor_sentiment_score'].value_counts().head(15))
```

```python
# Replace 'Not Discussed' with NaN, then coerce everything to numeric
df_em['crm_contractor_sentiment_score'] = df_em['crm_contractor_sentiment_score'].replace('Not Discussed', np.nan)
df_em['crm_contractor_sentiment_score'] = pd.to_numeric(df_em['crm_contractor_sentiment_score'], errors='coerce')

# Fill remaining NaN with the median of non-null values
median_score = df_em['crm_contractor_sentiment_score'].median()
df_em['crm_contractor_sentiment_score'] = df_em['crm_contractor_sentiment_score'].fillna(median_score)

print(f"Median score used for fill: {median_score}")
print(df_em['crm_contractor_sentiment_score'].describe())
```

Normalizing the crm_membership_level column - consolidating all variants into clean base tiers. Since counts for edge cases are very low (1-8), we merge them into the base tier (Gold, Silver, Bronze, etc.)

```python
print("Before cleaning:")
print(df_em['crm_membership_level'].value_counts())
```

```python
def normalize_membership(val):
    if pd.isna(val):
        return 'Not Discussed'
    val = val.strip().lower()
    
    # Not useful / not discussed
    if val in ['not discussed', 'yes', 'no', 'renewal complete', 'existing member',
               'accredited (implied by the context, but not explicitly stated)']:
        return 'Not Discussed'
    
    # In Progress
    elif val in ['in progress']:
        return 'In Progress'
    
    # Gold - includes all accredited-at-gold variants
    elif 'gold' in val:
        return 'Gold'
    
    # Silver - includes all accredited-at-silver variants
    elif 'silver' in val:
        return 'Silver'
    
    # Bronze - includes all accredited-at-bronze variants
    elif 'bronze' in val:
        return 'Bronze'
    
    # Accredited (plain, without tier) and suspended
    elif val in ['accredited', 'accredited (suspended)']:
        return 'Accredited'
    
    # Not Accredited
    elif val in ['not accredited']:
        return 'Not Accredited'
    
    # Members Only
    elif val in ['members only', 'member only', 'membership only', 'base membership', 'group membership']:
        return 'Members Only'
    
    # Standard - includes all standard variants
    elif 'standard' in val or val in ['membership (standard)']:
        return 'Standard'
    
    # Premier - includes all premier/prem variants
    elif 'premier' in val or val in ['prem', 'membership (premier)']:
        return 'Premier'
    
    # Express
    elif 'express' in val:
        return 'Express'
    
    # Band/Assisted patterns
    elif 'assisted' in val or 'band' in val:
        return 'Other'
    
    else:
        return 'Other'

df_em['crm_membership_level'] = df_em['crm_membership_level'].apply(normalize_membership)

print("\nAfter cleaning:")
print(df_em['crm_membership_level'].value_counts())
```

Cleaning the crm_auto_renewal_status column - converting to numeric with 0 = not set, 1 = active, 2 = other status

```python
print("Before cleaning:")
print(df_em['crm_auto_renewal_status'].value_counts(dropna=False))
```

```python
# Map text values and fill NaN
df_em['crm_auto_renewal_status'] = df_em['crm_auto_renewal_status'].replace({
    'Yes': '1', 'No': '0', 'Not Discussed': '0'
})
df_em['crm_auto_renewal_status'] = pd.to_numeric(df_em['crm_auto_renewal_status'], errors='coerce').fillna(0).astype(int)

print("After cleaning:")
print(df_em['crm_auto_renewal_status'].value_counts())
```

Cleaning numerical counts (like crm_agent_chase_count)

```python
if 'crm_agent_chase_count' in df_em.columns:
    # Standardize textual representations of counts
    df_em['crm_agent_chase_count'] = df_em['crm_agent_chase_count'].replace({
        'More than 5': '6', 
        'Multiple': '3'
    })
    # Convert to numeric, coerce errors for anything remaining
    df_em['crm_agent_chase_count'] = pd.to_numeric(df_em['crm_agent_chase_count'], errors='coerce').fillna(0).astype(int)

print(df_em['crm_agent_chase_count'].value_counts().head(10))
```

Final verification - checking remaining null values and dtypes after all cleaning

```python
print("Remaining null values:")
print(df_em.isnull().sum())
print(f"\nTotal rows: {len(df_em)}")
print(f"Total columns: {len(df_em.columns)}")
```

Giving type Category to categorical columns 

```python
cat_cols = [
    'Time_to_Renewal', 'crm_accreditation_completed', 'crm_timely_completion',
    'crm_progress_towards_accreditation', 'crm_delays_in_accreditation',
    'crm_contractor_suggested_leave', 'crm_contractor_engagement',
    'crm_contractor_sentiment', 'crm_dts_or_ssip_mentioned',
    'crm_customer_payment_intention', 'crm_competitors_mentioned',
    'crm_membership_level', 'crm_platform_issues_raised',
    'crm_agent_chased_contractor', 'crm_accreditation_issues',
    'crm_membership_overdue', 'crm_dissatisified_with_renewal_price',
    'crm_customer_complained', 'crm_refund_mentioned',
    'crm_negative_customer_experience', 'crm_dissatisfaction_with_support',
    'crm_financial_hardship_mentioned'
]

for col in cat_cols:
    df_em[col] = df_em[col].astype('category')
```
