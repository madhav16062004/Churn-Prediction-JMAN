Cleaning the data values, applying the binary logic

```python
def clean_binary_logic(val):
    if pd.isna(val): return 0
    
    # Standardize string (lowercase, remove asterisks and brackets)
    s = str(val).strip().lower().replace('*', '').replace('[', '').replace(']', '')
    
    # Map to 1 
    if s in ['yes', '1', '1.0', 'true']:
        return 1
    
    # Map everything else to 0 (Negative/No Signal)
    return 0

flag_columns = [
    'Serious_Complaint', 'Other_Complaint', 'Discussion_on_Price_Increase', 
    'Discount_or_Waiver_Requested', 'Call_Reschedule_Request', 
    'Explicit_Competitor_Mention', 'Explicit_Switching_Intent',
    'asked_for_discount', 'has_desired_to_cancel', 'Discount_Offered'
]

for col in flag_columns:
    if col in df_rc.columns:
        df_rc[col] = df_rc[col].apply(clean_binary_logic)
```

generalizing values in columns to no, yes

```python

# The Harmonizer Function
def harmonize_to_labels(val):
    if pd.isna(val) or str(val).strip() == "" or str(val).lower() == 'nan':
        return "Not Mentioned"
    
    # Clean formatting: remove **, [], and extra spaces
    s = str(val).strip().lower().replace('*', '').replace('[', '').replace(']', '')
    
    # Map to "Yes"
    if s in ['yes', '1', '1.0', 'true', 'won', 'renewed', 'renew']:
        return "Yes"
    
    # Map to "No"
    if s in ['no', '0', '0.0', 'false']:
        return "No"
    
    # If it's a specific category (like 'Price Increase'), keep it as is
    return str(val).strip()

# Identify all columns that contain "Binary-like" data
# We look for columns where the values (once cleaned) are mostly yes/no/0/1
cols_to_transform = []

for col in df_rc.columns:
    # Skip ID, Date, and numeric columns like Year or Call_Number
    if col in ['Co_Ref', 'Call_Date', 'Call_ID', 'Call_Year', 'Call_Number', 'Call_Length_Score']:
        continue
        
    # Get unique values, ignoring NaNs, and clean them
    unique_vals = df_rc[col].dropna().unique()
    clean_uniques = {str(v).strip().lower().replace('*', '').replace('[', '').replace(']', '') for v in unique_vals}
    
    # Identify if the column is primarily Yes/No/0/1 signals
    binary_signals = {'yes', 'no', '0', '1', '1.0', '0.0', 'true', 'false', 'not discussed', 'not applicable'}
    if clean_uniques.issubset(binary_signals) or len(clean_uniques) <= 3:
        cols_to_transform.append(col)

# Apply the transformation to all identified columns
for col in cols_to_transform:
    df_rc[col] = df_rc[col].apply(harmonize_to_labels)

# Final safety check: Fill any remaining NaNs in other text columns
df_rc = df_rc.fillna("Not Mentioned")

print(f"Cleaned {len(cols_to_transform)} columns to Yes/No/Not Mentioned format.")
print(f"Columns transformed: {cols_to_transform}")
```

Handling Nulls, NaNs etc for catgeorical columns 

```python
problem_cols = [
    'Renewal_Impact_Due_to_Price_Increase', 
    'Price_Switching_Mentioned', 
    'Competitor_Value_Comparison', 
    'Competitor_Benefits_Mentioned', 
    'Price_Range_Mentioned'
]

def final_cleanup(val):
    # Handle Nulls/Nans
    if pd.isna(val):
        return "Not Mentioned"
    
    # Standardize string for checking
    s = str(val).strip().lower()
    
    # If the value is a '0' variant, turn it into 'Not Mentioned'
    if s in ['0', '0.0', 'none', 'nan', 'not discussed', 'not applicable']:
        return "Not Mentioned"
    
    # For 'Price_Switching_Mentioned' and 'Impact', handle No variants
    # If you prefer these to stay as 'No' instead of 'Not Mentioned'
    if s == 'no':
        return "No"
    
    # Keep the actual categories/descriptions (e.g., "Better Value", "£500")
    return str(val).strip()

for col in problem_cols:
    if col in df_rc.columns:
        df_rc[col] = df_rc[col].apply(final_cleanup)
```

Cleaning up noise in the data

```python


# List of columns that still have messy sentences or noise
complex_cols = [
    'Renewal_Impact_Due_to_Price_Increase', 
    'Mentioned_Competitors', 
    'Price_Switching_Mentioned'
]

def final_label_cleaner(val):
    if pd.isna(val) or str(val).strip() == "":
        return "Not Mentioned"
    
    # 1. Standardize: lowercase and remove formatting symbols like ** or []
    s = str(val).strip().lower().replace('*', '').replace('[', '').replace(']', '')
    
    # 2. Priority: Identify "No" variants (even those with AI notes like 'No - Q9...')
    if s.startswith('no') or s in ['false', '0', '0.0']:
        return "No"
    
    # 3. Priority: Identify "Yes" variants (even long sentences starting with 'Yes')
    # Also catches values like "59 plus VAT" which imply a price was mentioned (a 'Yes' event)
    if s.startswith('yes') or s in ['true', '1', '1.0'] or any(char.isdigit() for char in s):
        return "Yes"
    
    # 4. Handle Noise (XXXX, UNAVAILABLE, punctuation, etc.)
    noise_values = ['not mentioned', 'not applicable', 'unknown', 'xxxx', 'unavailable', ':', '"', 'yes/no']
    if s in noise_values:
        return "Not Mentioned"
        
    # 5. Fallback: If it's a long string we don't recognize, it's likely a specific mention ('Yes')
    if len(s) > 5:
        return "Yes"
        
    return "Not Mentioned"

# Apply to your target columns
for col in complex_cols:
    if col in df_rc.columns:
        df_rc[col] = df_rc[col].apply(final_label_cleaner)

# Verification check
for col in complex_cols:
    print(f"Cleaned {col}: {df_rc[col].unique()}")
```

Combining columns with same values

```python
# 1. Define the columns you want to combine
# Even if they are mostly one value, we want to catch the outliers
interaction_cols = [
    'asked_for_discount', 
    'has_desired_to_cancel', 
    'Serious_Complaint', 
    'Other_Complaint',
    'Explicit_Switching_Intent'
]

# 2. Create a function to score the 'Intensity'
def calculate_interaction_score(row):
    score = 0
    # Add a point for every 'Positive' friction signal
    if row.get('asked_for_discount') == 'Yes': score += 1
    if row.get('has_desired_to_cancel') == 'Yes': score += 2  # Weighted higher because it's more serious
    if row.get('Serious_Complaint') == 'Yes': score += 2
    if row.get('Explicit_Switching_Intent') == 'Yes': score += 3
    return score

# 3. Apply the merge
df_rc['Call_Friction_Score'] = df_rc.apply(calculate_interaction_score, axis=1)

# 4. Now you can drop the original single-value columns to clean the noise
cols_to_drop = ['has_desired_to_cancel', 'asked_for_discount', 'Analysed_Call_Status', 'Analysed_Call']
df_rc = df_rc.drop(columns=[c for c in cols_to_drop if c in df_rc.columns])

print(f"Created 'Call_Friction_Score'. Dropped {len(cols_to_drop)} redundant columns.")
```

```python

# 1. Define the columns to handle
# Binary Flags: If 'Yes' appears once in any fragment, the whole year is 'Yes'
flag_cols = [
    'Serious_Complaint', 'Other_Complaint', 'Discussion_on_Price_Increase',
    'Discount_or_Waiver_Requested', 'Call_Reschedule_Request', 'Explicit_Competitor_Mention',
    'Explicit_Switching_Intent', 'Discount_Offered', 'asked_for_discount', 
    'has_desired_to_cancel', 'Has_Complaint', 'Has_Negative_Reaction'
]

# Categorical: Join multiple mentions with a semicolon (e.g., "Price; Login")
cat_cols = [
    'Complaint_Category', 'Customer_Reaction_Category', 'Agent_Renewal_Pitch_Category',
    'Customer_Renewal_Response_Category', 'Agent_Response_Category', 'Membership_Renewal_Decision',
    'Justification_Category', 'cancel_reason', 'discount_reason'
]

# 2. Helper function to aggregate text uniquely
def aggregate_text(series):
    # Remove 'Not Mentioned' and duplicates before joining
    unique_vals = [str(x) for x in series.unique() if str(x) not in ['Not Mentioned', 'Not Applicable', '0', 'nan']]
    return "; ".join(unique_vals) if unique_vals else "Not Mentioned"

# 3. Create the Aggregation Mapping
agg_map = {
    'Call_Date': 'max',          # Use the date of the most recent call
    'Call_Length_Score': 'mean',  # Average engagement
    'Call_Number': 'count'       # Total number of interactions/fragments
}

# Add Flags (use max: 'Yes' > 'No' > 'Not Mentioned')
for col in flag_cols:
    if col in df_rc.columns:
        agg_map[col] = 'max'

# Add Categories (Join strings)
for col in cat_cols:
    if col in df_rc.columns:
        agg_map[col] = aggregate_text

# 4. PERFORM THE GROUPING (The 'Squash')
# This creates 1 row per Customer (Co_Ref) per Year
df_rc_final = df_rc.groupby(['Co_Ref', 'Call_Year']).agg(agg_map).reset_index()

# 5. CREATE THE IMPACTFUL 'FRICTION SCORE'
# We merge those single-value/low-variance columns into one score
def calculate_friction(row):
    score = 0
    if row.get('Serious_Complaint') == 'Yes': score += 2
    if row.get('Explicit_Switching_Intent') == 'Yes': score += 3
    if row.get('Has_Complaint') == 'Yes': score += 1
    return score

df_rc_final['Interaction_Friction_Score'] = df_rc_final.apply(calculate_friction, axis=1)

print(f"Success! Data squashed from {len(df_rc)} rows to {len(df_rc_final)} unique customer-years.")
```
