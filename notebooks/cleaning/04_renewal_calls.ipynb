```python


from pathlib import Path


project_root = Path.cwd().parent.parent
data_path = project_root / "data" / "raw" / "renewal_calls.xlsx"


df_rc = pd.read_excel(data_path, sheet_name=0, header=0)

df_rc.head(10)
```

```python
# axis=1 means target columns
# how='all' means only drop if every value in the column is missing
df_rc = df_rc.dropna(axis=1, how='all')

# Verify the change
print(f"Columns remaining: {df_rc.columns.tolist()}")

```

```python
# This keeps only columns that don't start with the word 'Unnamed'
df_rc = df_rc.loc[:, ~df_rc.columns.str.contains('^Unnamed')]

df_rc.head()
```

```python


# Check null counts per column
print(df_rc.isnull().sum())
```

Handling NULL values for categorical columns 

```python
# List of categorical columns with high null counts
cat_cols = [
    'Churn_Category', 'Complaint_Category', 'Customer_Reaction_Category', 
    'Agent_Renewal_Pitch_Category', 'Customer_Renewal_Response_Category',
    'Mentioned_Competitors', 'Justification_Category', 'Reason_For_Renewal_Category'
]

# Fill with 'Not Mentioned' or 'None'
df_rc[cat_cols] = df_rc[cat_cols].fillna('Not Mentioned')
```

```python
df_rc.head(10)
```

```python
# List of indicator columns
flag_cols = [
    'Serious_Complaint', 'Other_Complaint', 'Discount_or_Waiver_Requested',
    'Call_Reschedule_Request', 'Explicit_Competitor_Mention', 'Explicit_Switching_Intent'
]

# Option A: Fill with 0 (Assuming Null means No)
df_rc[flag_cols] = df_rc[flag_cols].fillna(0)

# Option B: If they are strings like 'Yes'/NaN, map them:
df_rc['Serious_Complaint'] = df_rc['Serious_Complaint'].map({'Yes': 1}).fillna(0)
```

```python
# Create a clean 'discount_reason' column by extracting text inside brackets
# This looks for anything between ( and )
df_rc['discount_reason'] = df_rc['Discount_or_Waiver_Requested'].str.extract(r'\((.*?)\)')

# Fill the new column's nulls
df_rc['discount_reason'] = df_rc['discount_reason'].fillna('No Request')
```

```python
# If there is any value (even with numbers), mark as 1. If Null, mark as 0.
df_rc['asked_for_discount'] = df_rc['Discount_or_Waiver_Requested'].notnull().astype(int)
```

```python
# Remove digits and special characters from the reason
df_rc['discount_reason'] = df_rc['discount_reason'].str.replace(r'\d+', '', regex=True).str.strip()
```

```python
# Check null counts per column
print(df_rc.isnull().sum())
```

```python
# 1. Identify rows where the 'Call_Direction' is actually a long sentence
mask = df_rc['Call_Direction'].str.len() > 20

# 2. Fill in the gaps based on that sentence
# We can assume if they are 'renewing', the decision is 'Yes'
df_rc.loc[mask, 'Membership_Renewal_Decision'] = 'Renewed'
df_rc.loc[mask, 'Customer_Reaction_Category'] = 'Positive'

# 3. Fix the 'Call_Direction' column so it only contains 'Outbound' or 'Inbound'
# Since an agent was assisted, we'll label these as Outbound 
df_rc.loc[mask, 'Call_Direction'] = 'Outbound'

# 4. Fill the remaining empty ones 
df_rc['Call_Direction'] = df_rc['Call_Direction'].fillna('Unknown')

# Convert to lowercase, remove underscores, and strip extra spaces
df_rc['Call_Direction'] = df_rc['Call_Direction'].str.lower().str.replace('_', '').str.strip()

# Verify the result
print(df_rc['Call_Direction'].unique())
```

Analyze_Call column cleaning

```python
# Create a flag if the analyzed call mentions a discount
df_rc['discount_applied_flag'] = df_rc['Analysed_Call'].str.contains('discount', case=False, na=False).astype(int)
```

```python
import numpy as np

def categorize_analysis(val):
    val = str(val).strip()
    if val == 'nan' or val == 'NULL' or val == '':
        return 'Not Analysed'
    elif val.isdigit() and len(val) > 5:
        return 'ID/Numeric Record'
    elif 'Agent' in val or 'Customer' in val:
        return 'Detailed Summary'
    elif val in ['1', '3', 'Discounts/ incentives']:
        return 'Standard Category'
    else:
        return 'Other'

df_rc['Analysed_Call_Status'] = df_rc['Analysed_Call'].apply(categorize_analysis)
```

```python
# Fill the original column nulls so the code doesn't break later
df_rc['Analysed_Call'] = df_rc['Analysed_Call'].fillna('No Analysis Available')
```

Desire to cancel column cleaning

```python
import pandas as pd

# 1. Create the Boolean Flag
# We look for 'Desired to Cancel'. 
# Note: 'Undecided' will be False here. If you want Undecided to be True, 
# use: .str.contains('Desired to Cancel|Undecided', ...)
df_rc['has_desired_to_cancel'] = df_rc['Desire_To_Cancel'].str.contains('Desired to Cancel', case=False, na=False)

# 2. Extract the Reason (text inside brackets)
df_rc['cancel_reason'] = df_rc['Desire_To_Cancel'].str.extract(r'\((.*?)\)')

# 3. Clean up the Reason column
# Fill empty brackets/Nulls with 'No Reason Given'
df_rc['cancel_reason'] = df_rc['cancel_reason'].fillna('No Reason Given')

# If the original column had 'NAME' or 'Undecided' without brackets, 
# we label that as the reason so we don't lose that info.
df_rc.loc[df_rc['Desire_To_Cancel'].str.contains('NAME', na=False), 'cancel_reason'] = 'Personal Mention'
df_rc.loc[df_rc['Desire_To_Cancel'].str.contains('Undecided', na=False), 'cancel_reason'] = 'Undecided'

# 4. Verify
print(df_rc[['Desire_To_Cancel', 'has_desired_to_cancel', 'cancel_reason']].value_counts().head(10))
```

```python
# Standardize NULLs and 'Not Mentioned' into a single category
df_rc['Agent_Response_Category'] = df_rc['Agent_Response_Category'].fillna('Not Mentioned')

# If 'Not Mentioned' is already a string, this ensures they are identical
df_rc['Agent_Response_Category'] = df_rc['Agent_Response_Category'].replace(['NULL', 'nan', 'none'], 'Not Mentioned')

# Verify the result
print(df_rc['Agent_Response_Category'].value_counts())
```

```python
# 1. Fill actual NaN values first
df_rc['Agent_Response_Category'] = df_rc['Agent_Response_Category'].fillna('Not Mentioned')

# 2. Standardize string versions of 'Null' or 'NaN'
# We use a list of common "junk" strings found in Excel exports
junk_labels = ['null', 'nan', 'none', 'n/a', 'not mentioned', '']

df_rc['Agent_Response_Category'] = df_rc['Agent_Response_Category'].apply(
    lambda x: 'Not Mentioned' if str(x).strip().lower() in junk_labels else x
)

# 3. Quick check of your top 10 categories
print(df_rc['Agent_Response_Category'].value_counts().head(10))
```

Membership_Renewal_Decision column cleaning


```python


# 1. Fill actual NaNs with 'Pending/Unknown'
df_rc['Membership_Renewal_Decision'] = df_rc['Membership_Renewal_Decision'].fillna('Pending/Unknown')

# 2. Define a mapping dictionary to standardize the strings
# This handles the 'Yes/No' cases as 'Pending' to keep the model accurate
decision_map = {
    'Yes': 'Renewed',
    'No': 'Cancelled',
    'Yes/No': 'Pending/Unknown',
    'NULL': 'Pending/Unknown',
    'nan': 'Pending/Unknown'
}

# 3. Apply the mapping
# We use .get() so that any values already cleaned (like 'Renewed') stay as they are
df_rc['Membership_Renewal_Decision'] = df_rc['Membership_Renewal_Decision'].apply(
    lambda x: decision_map.get(x, x)
)

# 4. Verify the distribution
print(df_rc['Membership_Renewal_Decision'].value_counts())
```
