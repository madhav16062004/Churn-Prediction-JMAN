```python
import pandas as pd
import numpy as np
```

Using the Path library to read the file

```python


from pathlib import Path


project_root = Path.cwd().parent.parent
data_path = project_root / "data" / "raw" / "renewal_calls.csv"


df_rc = pd.read_csv(data_path, header=0)

df_rc.head(10)
```

This dataset contains call transcript metadata from renewal conversations, capturing the interactions between agents and customers during the critical subscription renewal window. 
It tracks customer sentiment and intent (such as desire to cancel or complaints), agent intervention strategies (like discount offers or specific renewal pitches), and competitive intelligence regarding rival pricing or benefits. 
The primary goal of this data is to serve as a predictor for post-renewal churn, allowing you to identify which "saved" customers are still at high risk of leaving in the future.

Drop columns that are completely null.

```python
# axis=1 means target columns
# how='all' means only drop if every value in the column is missing
df_rc = df_rc.dropna(axis=1, how='all')

# Verify the change
print(f"Columns remaining: {df_rc.columns.tolist()}")

```

Removing unnamed columns

```python
# This keeps only columns that don't start with the word 'Unnamed'
df_rc = df_rc.loc[:, ~df_rc.columns.str.contains('^Unnamed')]

df_rc.head()
```

check for null values for every column

```python
# Check null counts per column
print(df_rc.isnull().sum())
```

Handling NULL values for categorical columns 

```python
# List of categorical columns with high null counts
cat_cols = [
    'Churn_Category', 'Complaint_Category', 'Customer_Reaction_Category', 
    'Agent_Renewal_Pitch_Category', 'Customer_Renewal_Response_Category',
    'Mentioned_Competitors', 'Justification_Category', 'Reason_For_Renewal_Category'
]

# Fill with 'Not Mentioned' or 'None'
df_rc[cat_cols] = df_rc[cat_cols].fillna('Not Mentioned')
```

```python
df_rc.head(10)
```

Filling categorical column's null values


```python

flag_cols = [
    'Serious_Complaint', 'Other_Complaint', 'Discount_or_Waiver_Requested',
    'Call_Reschedule_Request', 'Explicit_Competitor_Mention', 'Explicit_Switching_Intent'
]

df_rc[flag_cols] = df_rc[flag_cols].fillna(0)

df_rc['Serious_Complaint'] = df_rc['Serious_Complaint'].map({'Yes': 1}).fillna(0)
```

cleaning discount_reason column derived from Discount_or_Waiver_Requested column to get the reason separately

```python
# Create a clean discount_reason column by extracting text inside brackets
df_rc['discount_reason'] = df_rc['Discount_or_Waiver_Requested'].str.extract(r'\((.*?)\)')

# Fill the new column's nulls
df_rc['discount_reason'] = df_rc['discount_reason'].fillna('No Request')
```

creating asked_for_discount to understand if the customer asked for a discount or not

```python
# If there is any value (even with numbers), mark as 1. If Null, mark as 0.
df_rc['asked_for_discount'] = df_rc['Discount_or_Waiver_Requested'].notnull().astype(int)
```

stripping the reason 

```python
# Remove digits and special characters from the reason
df_rc['discount_reason'] = df_rc['discount_reason'].str.replace(r'\d+', '', regex=True).str.strip()
```

checking the null values

```python
# Check null counts per column
print(df_rc.isnull().sum())
```

Cleaning the Call_Direction column and masking it based on conversation, integrating with Membership_Renewal_Decision and Customer_Reaction_Category.

```python
# Identify rows where the Call_Direction is actually a long sentence
mask = df_rc['Call_Direction'].str.len() > 20

# Fill in the gaps based on that sentence
# assuming if they are 'renewing', the decision is 'Yes'
df_rc.loc[mask, 'Membership_Renewal_Decision'] = 'Renewed'
df_rc.loc[mask, 'Customer_Reaction_Category'] = 'Positive'

# Fix the Call_Direction column so it only contains 'Outbound' or 'Inbound'
# Since an agent was assisted, label these as Outbound 
df_rc.loc[mask, 'Call_Direction'] = 'Outbound'

# Fill the remaining empty ones 
df_rc['Call_Direction'] = df_rc['Call_Direction'].fillna('Unknown')

# Convert to lowercase, remove underscores, and strip extra spaces
df_rc['Call_Direction'] = df_rc['Call_Direction'].str.lower().str.replace('_', '').str.strip()

print(df_rc['Call_Direction'].unique())
```

Analyze_Call column cleaning

```python
import numpy as np

def categorize_analysis(val):
    val = str(val).strip()
    if val == 'nan' or val == 'NULL' or val == '':
        return 'Not Analysed'
    elif val.isdigit() and len(val) > 5:
        return 'ID/Numeric Record'
    elif 'Agent' in val or 'Customer' in val:
        return 'Detailed Summary'
    elif val in ['1', '3', 'Discounts/ incentives']:
        return 'Standard Category'
    else:
        return 'Other'

df_rc['Analysed_Call_Status'] = df_rc['Analysed_Call'].apply(categorize_analysis)
```

```python
df_rc['Analysed_Call'] = df_rc['Analysed_Call'].fillna('No Analysis Available')
```

Desire to cancel column cleaning

```python

# look for Desired to Cancel
df_rc['has_desired_to_cancel'] = df_rc['Desire_To_Cancel'].str.contains('Desired to Cancel', case=False, na=False)

# Extract the Reason 
df_rc['cancel_reason'] = df_rc['Desire_To_Cancel'].str.extract(r'\((.*?)\)')

# Clean up the Reason column
# Fill empty brackets/Nulls with 'No Reason Given'
df_rc['cancel_reason'] = df_rc['cancel_reason'].fillna('No Reason Given')

# If the original column had 'NAME' or 'Undecided' without brackets, 
# we label that as the reason so we don't lose that info.
df_rc.loc[df_rc['Desire_To_Cancel'].str.contains('NAME', na=False), 'cancel_reason'] = 'Personal Mention'
df_rc.loc[df_rc['Desire_To_Cancel'].str.contains('Undecided', na=False), 'cancel_reason'] = 'Undecided'

print(df_rc[['Desire_To_Cancel', 'has_desired_to_cancel', 'cancel_reason']].value_counts().head(10))
```

Cleaning the Agent_Response_Category of nulls

```python
# Standardize NULLs and 'Not Mentioned' into a single category
df_rc['Agent_Response_Category'] = df_rc['Agent_Response_Category'].fillna('Not Mentioned')

# If 'Not Mentioned' is already a string, this ensures they are identical
df_rc['Agent_Response_Category'] = df_rc['Agent_Response_Category'].replace(['NULL', 'nan', 'none'], 'Not Mentioned')

print(df_rc['Agent_Response_Category'].value_counts())
```

```python

df_rc['Agent_Response_Category'] = df_rc['Agent_Response_Category'].fillna('Not Mentioned')
junk_labels = ['null', 'nan', 'none', 'n/a', 'not mentioned', '']

df_rc['Agent_Response_Category'] = df_rc['Agent_Response_Category'].apply(
    lambda x: 'Not Mentioned' if str(x).strip().lower() in junk_labels else x
)

print(df_rc['Agent_Response_Category'].value_counts().head(10))
```

Membership_Renewal_Decision column cleaning


```python


# Fill actual NaNs with 'Pending/Unknown'
df_rc['Membership_Renewal_Decision'] = df_rc['Membership_Renewal_Decision'].fillna('Pending/Unknown')

# Define a mapping dictionary to standardize the strings
# This handles the 'Yes/No' cases as 'Pending' to keep the model accurate
decision_map = {
    'Yes': 'Renewed',
    'No': 'Cancelled',
    'Yes/No': 'Pending/Unknown',
    'NULL': 'Pending/Unknown',
    'nan': 'Pending/Unknown'
}

df_rc['Membership_Renewal_Decision'] = df_rc['Membership_Renewal_Decision'].apply(
    lambda x: decision_map.get(x, x)
)

print(df_rc['Membership_Renewal_Decision'].value_counts())
```

Handling other catgeorical columns

```python
# Filling behavioral flags with 0
behavior_flags = [
    'Discussion_on_Price_Increase', 'Renewal_Impact_Due_to_Price_Increase',
    'Agent_Flagged_Membership_Status_Alert', 'Agent_Renewal_Initiation',
    'Price_Switching_Mentioned', 'Competitor_Value_Comparison',
    'Competitor_Benefits_Mentioned', 'Percentage_Price_Increase_Mentioned',
    'Monetary_Price_Increase_Mentioned', 'Price_Range_Mentioned',
    'Customer_Asked_For_Justification', 'Discount_Offered'
]
df_rc[behavior_flags] = df_rc[behavior_flags].fillna(0)

# Filling the new high-null categorical columns
new_cats = [
    'Customer_Response', 'Agent_Response_To_Cancel_Category',
    'Argument_That_Convinced_Customer_to_Stay_Category', 'Topic_Introduced_By'
]
df_rc[new_cats] = df_rc[new_cats].fillna('Not Mentioned')
```

Clubbing the fragmented data, each id has many columns in the data, fixing for better understanding and joining later

```python

# change text like "[Yes]" or "XXXX" into numbers/clean strings 
# so the 'max' logic works correctly

def clean_to_numeric(val):
    val = str(val).strip().lower()
    if val in ['yes', '1', '1.0', '[yes]', 'yes/no']:
        return 1
    return 0

# List of columns that indicate a "Yes/No" or "Presence" of something
flag_cols = [
    'asked_for_discount', 'has_desired_to_cancel', 'Serious_Complaint', 
    'Explicit_Competitor_Mention', 'Explicit_Switching_Intent', 
    'Discount_Offered', 'Discussion_on_Price_Increase'
]

for col in flag_cols:
    if col in df_rc.columns:
        df_rc[col] = df_rc[col].apply(clean_to_numeric)


# Do this BEFORE grouping so we know how many rows were originally there
df_rc['Call_Length_Score'] = df_rc.groupby(['Co_Ref', 'Call_Date'])['Call_ID'].transform('count')

# Logic: Use 'max' for flags (so we don't miss a 'Yes') 
# and 'last' for everything else (to get the final outcome of the call).

# Start with 'last' for every column
agg_dict = {col: 'last' for col in df_rc.columns if col not in ['Co_Ref', 'Call_Date']}

# Update specific columns to use 'max'
for col in flag_cols:
    if col in agg_dict:
        agg_dict[col] = 'max'

# Apply the grouping
df_rc = df_rc.groupby(['Co_Ref', 'Call_Date']).agg(agg_dict).reset_index()


print(f"Cleaning Complete.")
print(f"Total Unique Calls: {len(df_rc)}")
print(f"Total Columns Kept: {len(df_rc.columns)}")


print(df_rc[['Co_Ref', 'Call_Date', 'Call_Length_Score', 'Membership_Renewal_Decision']].head())
```

Cleaning of prefixes found

```python
import numpy as np

# A function to clean the prefixes and brackets
def clean_string_logic(val):
    val = str(val).strip().lower()
    if val in ['no', '[no]', 'prompt 3: no', 'prompt 4:', 'xxxx', 'unavailable', 'nan', '0', '0.0', '**no**']:
        return 'No'
    if val in ['yes', '[yes]', '1', '1.0']:
        return 'Yes'
    if 'yes/no' in val or 'not discussed' in val or 'not applicable' in val:
        return 'Unknown'
    return val.capitalize()

flag_cols = ['Explicit_Competitor_Mention', 'Explicit_Switching_Intent', 'Serious_Complaint', 'Discussion_on_Price_Increase']
for col in flag_cols:
    df_rc[col] = df_rc[col].apply(clean_string_logic)
```

Giving type Category to categorical columns 

```python
cat_cols = [
    'Call_Direction', 'Churn_Category', 'Complaint_Category', 
    'Customer_Reaction_Category', 'Agent_Renewal_Pitch_Category',
    'Membership_Renewal_Decision', 'Analysed_Call_Status'
]

for col in cat_cols:
    df_rc[col] = df_rc[col].astype('category')
```
